```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

df = pd.read_csv("raman_data.CSV")
df
```

```python
df_filter1 = df[df["X"] > 20]

x = np.array(df_filter1["X"].tolist())
y = np.array(df_filter1["Y"].tolist())

peak, _ = find_peaks(y, height=15000,distance=100)
plt.plot(x[peak], y[peak], 'ro')
plt.axvline(x=x[peak][0],color='r',linestyle='--', linewidth=1,label=str(x[peak][0]))
plt.axvline(x=x[peak][1],color='g',linestyle='--', linewidth=1,label=str(x[peak][1]))
plt.plot(x,y)
plt.ylabel("Intensity")
plt.xlabel("Raman Shift[1/cm]")
plt.legend()
plt.show()
```

```python
df_filter2 = df[df["X.1"] > 15]
x2 = np.array(df_filter2["X.1"].tolist())
y2 = np.array(df_filter2["Y.1"].tolist())

peak2, _ = find_peaks(y2, height=4000,distance=100)
plt.plot(x2[peak2], y2[peak2], 'ro')
plt.axvline(x=x2[peak2][0],color='r',linestyle='--', linewidth=1,label=str(x2[peak2][0]))
plt.axvline(x=x2[peak2][1],color='g',linestyle='--', linewidth=1,label=str(x2[peak2][1]))
plt.plot(x2,y2)
plt.ylabel("Intensity")
plt.xlabel("Raman Shift[1/cm]")
plt.legend()
plt.show()
```
