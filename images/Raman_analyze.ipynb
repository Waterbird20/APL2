```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

df = pd.read_csv("raman_data.CSV")
df
```

```python
df_filter1 = df[df["X"] > 20]

x = np.array(df_filter1["X"].tolist())
y = np.array(df_filter1["Y"].tolist())

peak, _ = find_peaks(y,height=2500,width=20)
plt.plot(x[peak],y[peak],'ro')
plt.axvline(x=x[peak][0],color='r',linestyle='--', linewidth=1,label=str(x[peak][0]))
plt.axvline(x=x[peak][1],color='g',linestyle='--', linewidth=1,label=str(x[peak][1]))
plt.axvline(x=x[peak][2],color='m',linestyle='--', linewidth=1,label=str(x[peak][2]))
plt.axvline(x=x[peak][3],color='y',linestyle='--', linewidth=1,label=str(x[peak][3]))
plt.plot(x,y)
plt.ylabel("Intensity")
plt.xlabel("Raman Shift[1/cm]")
plt.title("Reduced Graphene Oxide")
plt.legend()
plt.savefig('rgo.png', facecolor='#eeeeee',
            bbox_inches='tight', pad_inches=0.3)
plt.show()
```

```python
df_filter2 = df[df["X.1"] > 15]
x2 = np.array(df_filter2["X.1"].tolist())
y2 = np.array(df_filter2["Y.1"].tolist())

peak2, _ = find_peaks(y2, height=4000,distance=100)
plt.plot(x2[peak2], y2[peak2], 'ro')
plt.axvline(x=x2[peak2][0],color='r',linestyle='--', linewidth=1,label=str(x2[peak2][0]))
plt.axvline(x=x2[peak2][1],color='g',linestyle='--', linewidth=1,label=str(x2[peak2][1]))
plt.plot(x2,y2)
plt.ylabel("Intensity")
plt.xlabel("Raman Shift[1/cm]")
plt.title("Graphite")
plt.legend()
plt.savefig('gr_.png', facecolor='#eeeeee',
            bbox_inches='tight', pad_inches=0.3)
plt.show()
```

```python
df_filter3 = df[df["X.2"] > 30]
x3 = np.array(df_filter3["X.2"].tolist())
y3 = np.array(df_filter3["Y.2"].tolist())

peak3, _ = find_peaks(y3, height=20000,distance=100)
plt.plot(x3[peak3], y3[peak3], 'ro')
plt.axvline(x=x3[peak3][0],color='r',linestyle='--', linewidth=1, label=str(x3[peak3][0]))
plt.axvline(x=x3[peak3][1],color='g',linestyle='--', linewidth=1, label=str(x3[peak3][1]))
plt.plot(x3,y3)
plt.title("Graphene Oxide")
plt.ylabel("Intensity")
plt.xlabel("Raman Shift[1/cm]")
plt.legend()
plt.savefig('go_.png', facecolor='#eeeeee',
            bbox_inches='tight', pad_inches=0.3)
plt.show()
```

```python
rgo_x = np.array(df_filter3["X"].tolist())
rgo_y = np.array(df_filter3["Y"].tolist())
gr_x = np.array(df_filter3["X.1"].tolist())
gr_y = np.array(df_filter3["Y.1"].tolist())
go_x = np.array(df_filter3["X.2"].tolist())
go_y = np.array(df_filter3["Y.2"].tolist())

plt.plot(rgo_x,rgo_y,label='RGO')
plt.plot(gr_x,gr_y,label='GR')
plt.plot(go_x,go_y,label='GO')

plt.axvline(x=1580,color='r',linestyle='--', linewidth=1)
plt.axvline(x=1350,color='r',linestyle='--', linewidth=1)
plt.axvline(x=2710,color='r',linestyle='--', linewidth=1)
plt.axvline(x=2910,color='r',linestyle='--', linewidth=1)

plt.text(1570,36000,"1580",weight='bold',color='r',ha='right',rotation=90)
plt.text(1560,42000,"G",weight='bold',ha='right',size=12)
plt.text(1340,36000,"1350",weight='bold',color='r',ha='right',rotation=90)
plt.text(1330,42000,"D",weight='bold',ha='right',size=12)
plt.text(2700,36000,"2710",weight='bold',color='r',ha='right',rotation=90)
plt.text(2700,42000,"2D",weight='bold',ha='right',size=12)
plt.text(2900,36000,"2910",weight='bold',color='r',ha='right',rotation=90)
plt.text(2950,42000,"D+D'",weight='bold',ha='center',size=12)

plt.ylim(0,45000)
plt.ylabel("Intensity",size=12)
plt.xlabel("Raman Shift[1/cm]",size=12)
plt.legend()
plt.savefig('graphene_vibration mode.png', facecolor='#eeeeee',
            bbox_inches='tight', pad_inches=0.3)
plt.show()
```
